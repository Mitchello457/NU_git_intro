```python
# import needed modules
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# get data from file
energy = pd.read_csv('data/energy.sph',delim_whitespace = True, comment=None,header = None)
```

```python
# make plots
f,(ax1,ax2,ax3,ax4) = plt.subplots(4,sharex = True, figsize = (8,12))

# plot data
## potential

ax1.plot(energy[0],energy[1],'k')
ax1.set_title('Energy from energy.sph',size=16,name='sans-serif',va='bottom',weight='semibold')
ax1.set_ylabel('Potential')

## kinetic

ax2.plot(energy[0],energy[2],'k')
ax2.set_ylabel('Kinetic')

## internal

ax3.plot(energy[0],energy[3],'k')
ax3.set_ylabel('Internal')

## total

ax4.plot(energy[0],energy[4],'k')
ax4.set_ylabel('Total')

# adjust spacing
f.subplots_adjust(hspace=0)
```
